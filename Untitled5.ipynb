```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# reading the dataset

```python
iris = 'iris.xls'
df = pd.read_excel(iris)
```

# visualizations

# 1 box plot

```python
plt.figure(figsize=(20,10))
sns.boxplot(x='SL',y='SW',hue='Classification',data=df)
plt.title('Sepal length vs Sepal width',size=20)
plt.figure(figsize=(20,10))
sns.boxplot(x='PL',y='PW',hue='Classification',data=df)
plt.title('Petal length vs Petal width',size=20)
```

# 2 violin plot

```python
plt.figure(figsize=(15,15))
sns.violinplot(x='Classification',y='PL',data=df)
plt.title('Petal length of different Classification')

```

# 3 strip plot

```python
plt.figure(figsize=(15,15))
sns.stripplot(x='Classification',y='SL',data=df)
plt.title('Sepal length of different Classification')
```

# 4 swarm plot

```python
plt.figure(figsize=(15,15))
sns.swarmplot(x='Classification',y='SW',data=df,hue='Classification')
plt.title('Sepal width of different Classification')
```
